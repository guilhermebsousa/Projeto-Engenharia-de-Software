```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


fig, ax = plt.subplots(figsize=(16, 6))

# Etapas do fluxo do usuário
steps = [
    "Onboarding",
    "Login/Cadastro",
    "Cadastro de Prod.",
    "Monit. de Estoque",
    "Sugestão de Repos.",
    "Conf.de Pedido",
    "Histórico"
]

x_positions = range(len(steps))
y_position = 0


box_width = 10.5
box_height = 0.5
spacing = 2.0
for i, step in enumerate(steps):
    box = mpatches.FancyBboxPatch(
        (i * (box_width + spacing), y_position), box_width, box_height,
        boxstyle="round,pad=0.1",
        edgecolor="black",
        facecolor="#a1d6ff"
    )
    ax.add_patch(box)
    ax.text(i * (box_width + spacing) + box_width/2, y_position + box_height/2, step,
            ha="center", va="center", fontsize=10, weight="bold")


for i in range(len(steps) - 1):
    start_x = i * (box_width + spacing) + box_width
    end_x = (i + 1) * (box_width + spacing)
    ax.annotate(
        "",
        xy=(end_x, y_position + box_height/2),
        xytext=(start_x, y_position + box_height/2),
        arrowprops=dict(arrowstyle="->", lw=2)
    )


ax.set_xlim(-0.5, len(steps) * (box_width + spacing))
ax.set_ylim(-0.5, 2)
ax.axis("off")

plt.title("Fluxo do usuário no app", fontsize=20, weight="bold")
plt.show()
```
